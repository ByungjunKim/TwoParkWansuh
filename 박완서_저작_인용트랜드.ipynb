```python
df = pd.read_pickle('../data/240731_2008_2022_한국현대문학_통합_성별_생년_v2.pkl')
df
```

```python
park = pd.read_excel('./2008_2022_한국현대문학_인문_복합학_추가데이터_참고문헌_박완서_중복삭제_제목통일작업.xlsx',sheet_name='시트1')
park
```

```python
# 새로운 컬럼 생성
def categorize_year(year):
    if 1970 <= year < 1980:
        return '70s'
    elif 1980 <= year < 1990:
        return '80s'
    elif 1990 <= year:
        return '90_00s'

park['decade'] = park['h.labeled_title_published.year'].apply(categorize_year)
park['decade'].value_counts()
```

```python
# df와 park을 article-id와 artiId를 기준으로 조인
park = pd.merge(df, park, left_on='article-id', right_on='artiId', how='right')
```

```python
len(park.loc[pd.isna(park['pub-year'])])
```

```python
park = park.loc[~pd.isna(park['pub-year'])].reset_index(drop=True)
```

```python
park['pub-year'] = park['pub-year'].astype(int)
```

```python
park.to_excel('./2008_2022_한국현대문학_인문_복합학_추가데이터_참고문헌_박완서_중복삭제_제목통일작업_게재연도_성별.xlsx')
```

```python
park['pub-year'].value_counts().sort_index()
```

```python
# 시계열 트렌드 분석
trend = park.groupby(['pub-year', 'decade']).size().unstack().fillna(0)

# 시각화
plt.figure(figsize=(12, 8))  # 고화질을 위해 그림 크기 설정
trend.plot(kind='bar', stacked=True)
plt.title('연도별 박완서 저작시대 트랜드', fontsize=15)
plt.xlabel('Publication Year', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.legend(title='Decade', fontsize=10)
plt.xticks(rotation=45, fontsize=10)  # x축 범례 45도 회전
plt.yticks(fontsize=10)
plt.show()
```

```python
# 남성 데이터
park_male = park[park['gender'] == '남']
trend_male = park_male.groupby(['pub-year', 'decade']).size().unstack().fillna(0)
# 모든 연도를 포함하도록 재구성
all_years = pd.DataFrame(index=range(park['pub-year'].min(), park['pub-year'].max() + 1))
trend_male = all_years.join(trend_male, how='left').fillna(0)

# 여성 데이터
park_female = park[park['gender'] == '여']
trend_female = park_female.groupby(['pub-year', 'decade']).size().unstack().fillna(0)

# 시각화
plt.figure(figsize=(12, 8))  # 고화질을 위해 그림 크기 설정
trend_male.plot(kind='bar', stacked=True)
plt.title('연도별 박완서 저작시대 트랜드 (남성)', fontsize=15)
plt.xlabel('Publication Year', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.legend(title='Decade', fontsize=10)
plt.xticks(rotation=45, fontsize=10)  # x축 범례 45도 회전
plt.yticks(fontsize=10)
plt.show()

plt.figure(figsize=(12, 8))  # 고화질을 위해 그림 크기 설정
trend_female.plot(kind='bar', stacked=True)
plt.title('연도별 박완서 저작시대 트랜드 (여성)', fontsize=15)
plt.xlabel('Publication Year', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.legend(title='Decade', fontsize=10)
plt.xticks(rotation=45, fontsize=10)  # x축 범례 45도 회전
plt.yticks(fontsize=10)
plt.show()
```

```python
# 엑셀 파일로 내보내기
with pd.ExcelWriter('./분석결과/박완서_저작시대_인용_트랜드_성별.xlsx') as writer:
    trend_male.to_excel(writer, sheet_name='Trend_Male')
    trend_female.to_excel(writer, sheet_name='Trend_Female')
```
